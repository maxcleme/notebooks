```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import matplotlib as mpl
import pandas as pd
import matplotlib.colors as c

palettes = [
    'hama.csv',
    'nabbi.csv',
    'perler.csv',
    'perler_mini.csv',
    'artkal_a.csv',
    'artkal_c.csv',
    'artkal_r.csv',
    'artkal_s.csv'
]

fig, axs = plt.subplots(2, 4, figsize=(20, 8), subplot_kw=dict(projection='polar'))
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()

for i in range(len(palettes)):
    palette = palettes[i]
    ax=axs[i]
    
    df = pd.read_csv(palette)
    df['color_tuple'] = list(zip(df['rgb_r']/256, df['rgb_g']/256, df['rgb_b']/256))
    df['color_tuple'] = df['color_tuple'].map(lambda x: c.to_rgb(x))
    
    # Compute areas and colors
    r = df['hsl_s']
    theta = df['hsl_h'] * np.pi / 180
    colors = df['color_tuple']
    
    # Draw
    area = 100
    ax.set_title(palette.split('.')[0])
    ax.set_theta_zero_location("N")
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_axis_off()
    ax.grid(False)
    ax.scatter(theta, r, c=colors, s=area, alpha=0.75)    


plt.show() # Replace with plt.savefig if you want to save a file
```
